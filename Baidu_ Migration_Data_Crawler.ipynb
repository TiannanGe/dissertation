```python
# -*- coding: utf-8 -*-
import requests 
import json 
import time
import xlrd
import xlwt
from ChineseAdminiDivisionsDict import CitiesCode, ProvinceCode

def migration_all_date(areaname,classname,no,direction): #Definition generates different periods, different cities, different migration directions
    if no == -1 :
        no = CitiesCode[str(areaname)]
# Create a workbook
    workbook = xlwt.Workbook(encoding = 'utf-8')    # Creating a workbook Setting the code
    worksheet = workbook.add_sheet('Sheet', cell_overwrite_ok=True)    # Create a worksheet
# Write the code of each city and its name in the header of the line
    if direction == 'in' :
        nameofdire = '迁入来源地'
    if direction == 'out':
        nameofdire = '迁出目的地'
    CitiesOrder = {}                         #Empty dictionary for storing city serial numbers
    worksheet.write(0 , 0 , label='城市代码')                     
    worksheet.write(0 , 1 , label=str(nameofdire))                 
    times = 1
    for key , value in CitiesCode.items():
        worksheet.write(times , 0 , label=str(value))                
        worksheet.write(times , 1 , label=str(key))                 
        CitiesOrder[str(key)] = times                           
        times += 1
# Set date
    datelist = []                                    #Date List
    counter_data = 2                                 #date counter
    for date1 in range(20230201,20230232):           
        datelist.append(date1)
    for date in datelist:                            #Iterate over all dates
        datename = date
        time.sleep(1)  
        url=f'http://huiyan.baidu.com/migration/cityrank.jsonp?dt={classname}&id={no}&type=move_{direction}&date={date}'
        print(url)
        response=requests.get(url, timeout=100) #Make a request and jsonize it
        time.sleep(1)  
        r=response.text[4:-1] #trim the head and tail
        data_dict=json.loads(r) #characterization
        if data_dict['errmsg']=='SUCCESS':
            data_list=data_dict['data']['list']
            time.sleep(1)
            ################写入###############
            worksheet.write(0 , counter_data , label=datename)             #Write to table header ---- date
            for a in range(len(CitiesCode)):
                worksheet.write(a+1 , counter_data , label=0)              #First set all city values in the column to 0 for the current date
            ############获取数据###########
            for i in range (len(data_list)):
                city_name=data_list[i]['city_name'] 
                value=data_list[i]['value']         
            ##############写入#############
                worksheet.write(CitiesOrder[str(city_name)] , counter_data , label=value)#Look up the city serial number dictionary and write the corresponding value in the 
            counter_data += 1                                 #Date counter plus one
    workbook.save(f"{areaname}-{nameofdire}.xls")      #save


def circu_exe_direction(areaname,classname,no):
    mukous = ['in','out']
    for mukou in mukous:
        migration_all_date(areaname,classname,no,mukou)
    print(str(areaname)+'---','Done.')


if __name__=="__main__":
    circu_exe_direction('儋州市','city',-1)
    print('All Done.')
```
